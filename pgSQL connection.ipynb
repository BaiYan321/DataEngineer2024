```python
import pandas as pd
import os.path
```

```python
# !pip install sqlalchemy psycopg2-binary

from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi') #( user name: password @ host: port / database schema)
# need to create user name "root" with the password "root" and database "ny_taxi" in pgAdmin
```

```python
# if it works, we connect to the localhost database
engine.connect()
```

```python
#import text
from sqlalchemy import text

#写下一个query
query = '''
SELECT 1
'''

# 方法一
pd.DataFrame(engine.connect().execute(text(query)))
```

```python
# 方法二
pd.read_sql_query(sql=text(query), con=engine.connect())
```

```python
# 这个是读取数据库里各种信息的方式
query = """
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog'
AND schemaname != 'information_schema'
"""
pd.read_sql_query(sql=text(query), con=engine.connect())
# it shows that the database is clean, no schema
```

```python
#check the file if exist, if exist, read the file, otherwise, download it and read it
url = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/yellow_tripdata_2021-01.csv.gz"
filename = url.split('/')[-1]
#print(filename)

if os.path.exists(filename): #if exist
    df = pd.read_csv(filename, compression='gzip', nrows = 100)
else:
    !wget url
    df = pd.read_csv(filename, compression='gzip', nrows = 100)
df 
```

```python
# convert the data type to datetime
df.tpep_pick_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
# 把这个表，写入到postgre里
# 注意要强调哪个 schema name
df.to_sql(name='yellow_tripdata_trip', con=engine, schema='public', index = False, if_exists='append')

query = """
SELECT *
FROM pg_catalog.pg_tables
WHERE schemaname != 'pg_catalog'
AND schemaname != 'information_schema'
"""

pd.read_sql_query(sql=text(query), con=engine.connect())
# 把df写入到schema里
```
